# Importing the Libraries

```python
# Importing the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import operator
from sklearn.model_selection import train_test_split 
%matplotlib inline
```

# Importing the dataset

```python
# Importing the dataset by using pandas

dataset1 = pd.read_csv("data.csv")
dataset1.head()
```

```python
dataset1.info()
```

```python
dataset1.describe()
```

```python
drop_cols = dataset1.columns[28:54]
dataset = dataset1.drop(drop_cols, axis=1)
dataset_f = dataset.drop(columns=["Photo","Nationality","Flag","Club Logo","Preferred Foot","Loaned From","Contract Valid Until"],axis=1)

dataset_f.head()
```

```python
dataset_f.isnull().sum()
```

```python
dataset_f = dataset_f.fillna(dataset_f.median())
dataset_f.isnull().sum()
```

```python
columns = ["Club","Work Rate","Body Type","Real Face","Position","Joined","Height","Weight","Release Clause"]

for i in columns:
    dataset_f[i].fillna(dataset_f[i].value_counts().idxmax(), inplace=True)
```

```python
dataset_f.isnull().sum()
```

```python
category = ["Name","Club","Value","Wage","Work Rate","Body Type","Real Face","Position","Joined","Height","Weight","Release Clause"]
for i in category:
    dataset_f[i] = pd.factorize(dataset_f[i])[0]
```

```python
dataset_f.info()
```

```python
dataset_f.columns[4]
```

```python
X = dataset_f.drop("Overall",1)
y = dataset_f['Overall']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
lin_regressor.fit(X_train,y_train)
prediction_linear=lin_regressor.predict(X_test)
```

```python
from sklearn.linear_model import Lasso
lasso_regressor=Lasso()
lasso_regressor.fit(X_train,y_train)
prediction_lasso=lasso_regressor.predict(X_test)
```

```python
from sklearn.linear_model import Ridge
ridge_regressor=Ridge()
ridge_regressor.fit(X_train,y_train)
prediction_ridge=ridge_regressor.predict(X_test)
```

```python
from sklearn.linear_model import ElasticNet
model_enet = ElasticNet()
model_enet.fit(X_train, y_train) 
prediction_elasticnet= model_enet.predict(X_test)
```

```python
print("Linear")
df = pd.DataFrame({'Actual': y_test, 'Predicted': prediction_linear})
df.head()
```

```python
print("Ridge")
df = pd.DataFrame({'Actual': y_test, 'Predicted': prediction_ridge})
df.head()
```

```python
print("Lasso")
df = pd.DataFrame({'Actual': y_test, 'Predicted': prediction_lasso})
df.head()
```

```python
print("Elastic Net")
df = pd.DataFrame({'Actual': y_test, 'Predicted': prediction_elasticnet})
df.head()
```

```python
from sklearn import metrics
```

```python
print("Linear Regression")
mae_linear=metrics.mean_absolute_error(y_test, prediction_linear)
print('Mean Absolute Error:',mae_linear )   
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction_linear)) 
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction_linear)))
```

```python
print("Ridge Regression")
mae_ridge=metrics.mean_absolute_error(y_test, prediction_ridge)
print('Mean Absolute Error:',mae_ridge )   
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction_ridge)) 
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction_ridge)))
```

```python
print("Lasso Regression")
mae_lasso=metrics.mean_absolute_error(y_test,prediction_lasso)
print('Mean Absolute Error:',mae_lasso)
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction_lasso))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction_lasso)))
```

```python
print("Elastic_net Regression")
mae_elasticnet=metrics.mean_absolute_error(y_test, prediction_elasticnet)
print('Mean Absolute Error:',mae_elasticnet)
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction_elasticnet))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction_elasticnet)))
```

```python
plt.figure(figsize=(12,8))
plt.plot(["Linear","Ridge","Lasso","ElasticNet"],[mae_linear,mae_ridge,mae_lasso,mae_elasticnet],marker='o')
plt.grid()
plt.xlabel("Types of Regularization")
plt.ylabel("Mean Absolute Error")
```
