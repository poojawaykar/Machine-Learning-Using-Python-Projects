# Importing the Libraries

```python
# Importing the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import operator
```

# Importing the dataset

```python
# Importing the dataset by using pandas

dataset1 = pd.read_csv("Loan.csv")
dataset1
```

```python
dataset1.info()
```

```python
dataset1.describe()
```

```python
dataset = dataset1.drop("Loan_ID",1)
dataset.info()
```

```python
dataset["Dependents"] = dataset["Dependents"].replace(to_replace = "3+", value = "3")
dataset["Dependents"] = dataset["Dependents"].astype(object)

dataset
```

```python
dataset.isnull().sum()
```

```python
median1 = dataset["LoanAmount"].median()
median2 = dataset["Loan_Amount_Term"].median()
median3 = dataset["Credit_History"].median()

dataset["LoanAmount"].replace(np.nan,median1,inplace=True)
dataset["Loan_Amount_Term"].replace(np.nan,median1,inplace=True)
dataset["Credit_History"].replace(np.nan,median1,inplace=True)
```

```python
mode1 = dataset["Gender"].mode().values[0]
mode2 = dataset["Married"].mode().values[0]
mode3 = dataset["Dependents"].mode().values[0]
mode4 = dataset["Self_Employed"].mode().values[0]

dataset["Gender"] = dataset["Gender"].replace(np.nan,mode1)
dataset["Married"] = dataset["Married"].replace(np.nan,mode1)
dataset["Dependents"] = dataset["Dependents"].replace(np.nan,mode1)
dataset["Self_Employed"] = dataset["Self_Employed"].replace(np.nan,mode1)
```

```python
dataset.isnull().sum()
```

```python
duplicate = dataset.duplicated()
print(duplicate.sum())
```

```python
category = ["Gender","Married","Dependents","Education","Self_Employed","Property_Area","Loan_Status"]
for i in category:
    dataset[i] = pd.factorize(dataset[i])[0]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
X = dataset.iloc[:,:-1]
X.head()
```

```python
y = dataset.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# Support Vector Machines

```python
from sklearn.svm import SVC

model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
